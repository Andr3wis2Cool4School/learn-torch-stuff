```python
def z_score(series):
    _mean = series.sum()/series.count()
    print(_mean)
    
    std = (((series - _mean)**2).sum() / (series.count() - 1)) ** 0.5
    
    print(std)
    
    new_series = (series - _mean) / std
    
    return new_series
```

```python
df['综合能力'] = z_score(df['综合能力'])
```

```python
X = df
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
import torch
import torch.nn as nn
```

```python
SalaryNet = SalaryNet(13, 100, 20, 1)
```

```python
criterion = nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(SalaryNet.parameters(), 1e-5)
```

```python
import math 
import time
import numpy as np
```

```python
batch_size = 512
epoch = 50

for layer in SalaryNet.modules():
    if isinstance(layer, nn.Linear):
        nn.init.xavier_uniform_(layer.weight)
        
loss_holder = []

blocks = math.ceil(y_train.count()/batch_size)

# setting loss as inf 
loss_value = np.inf

step = 0
for i in range(epoch):
    train_count = 0
    batchs = 0
    for j in range(blocks):
        train_x_data = torch.Tensor(X_train.iloc[j*batch_size: (j+1)*batch_size].values)
        train_x_data.requires_grid = True
        
        train_y_data = torch.Tensor(y_train.iloc[j*batch_size: (j+1)*batch_size].values)
        
        out = SalaryNet(train_x_data)
        loss = criterion(out.squeeze(1), train_y_data)
        
        optimizer.zero_grad()
        
        loss.backward()
        
        optimizer.step()
        
        msg = 'epoch: {}, Train Loss: {:.6f}, Dealed/Records: {}/{}'.format(i, loss/batch_size, (j+1)*batch_size,y_train.count())
        
        print(msg)
```
